```python
import matplotlib.pyplot as plt

# Create dataset
ages = [16, 22, 25, 32, 35, 40, 35, 50, 45, 60,
        23, 37, 32, 37, 42, 27, 52, 27, 62, 28]

#create the histogram
plt.figure()
plt.hist(ages,bins=6)

#add labels and tittle
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.title("Age distribution of a population")

#display the histogram
plt.show()
```
